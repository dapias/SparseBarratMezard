```julia
cav1 = load("rho_cavity.jld")
cav2 = load("rho_c2.jld");
```

```julia
rho_cav = cav1["rho"]
l_grid = cav1["lambdas"];
```

```julia
rho_cav2 = cav2["rho"]
l_grid2 = cav2["lambdas"];
epsilon = cav2["epsilon"];
```

```julia
plot(l_grid, rho_cav,  ls = :dash, label = L"\epsilon = 0.005im")
plot!(l_grid2, rho_cav2, ls = :dash, label = L"\epsilon"*"= $(epsilon)")
```
